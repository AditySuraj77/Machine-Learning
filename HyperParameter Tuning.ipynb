```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
pd.set_option('display.max_columns', None)
```

```python
# Working with dataset
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
data_frames = pd.DataFrame(breast_cancer_dataset.data, columns= breast_cancer_dataset.feature_names)
data_frames.head()
```

```python
data_frames['Label'] = breast_cancer_dataset.target
data_frames.head()
```

```python
data_frames.shape
```

```python
data_frames.isnull().sum()
```

```python
data_frames['Label'].value_counts()

#1 --> Benign

#0 --> Malignant
```

```python
# Define the features var & target var
X = data_frames.drop(columns='Label', axis=1)
Y = data_frames['Label']
```

```python
print(X)
print(Y)
```

```python
#Convert the target & feature in Numpy Arrays
X = np.asarray(X)
Y = np.asarray(Y)
```

```python
print(X)
print(Y)
```

# GridSearchCV

```python
# Selecting model
model = SVC()
```

```python
# Define the Prarameters
parameters = {'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
              'C':[1,5,10,20]}
```

```python
#gridsearch 
classifier = GridSearchCV(model,parameters,cv=5)
```

```python
#Fit the data X & Y
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
classifier.best_score_ # model gives best average best_accuracy
```

```python
classifier.best_params_ # model gives best parameters
```

```python
classifier_dataframes = pd.DataFrame(classifier.cv_results_) # Converting into Dataframes a model hyperparamenter process
classifier_dataframes.head()
```

```python
grid_search_result = classifier_dataframes[['param_C','param_kernel','mean_test_score']]
grid_search_result
```

# Final Result

# ----------------------------------------------------------------

## In GridSearchCV

# ----------------------------------------------------------------

### Best Model Accueracy is 0.9525694767893185
### Best Model Parameters is {'C': 10, 'kernel': 'linear'}
