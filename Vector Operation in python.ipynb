```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,5,5)
plt.show()
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1)
plt.xlim=(-8,8)
plt.ylim=(-8,8)
plt.show()
```

```python
plt.quiver(0,0,2,3,scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim=(-8,8)
plt.ylim=(-8,8)
plt.show()
```

```python
plt.quiver(0,0,2,3,scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-3,-2,scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim=(-8,8)
plt.ylim=(-8,8)
plt.show()
```

# Addition of two vector

```python
vec_1 = np.asarray([0,0,2,3])
vec_2 = np.asarray([0,0,3,-2])
sum= vec_1 + vec_2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2,scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,5,1,scale_units='xy', angles='xy', scale=1, color='g')

plt.xlim=(-8,8)
plt.ylim=(-8,8)
plt.show()
```

# Subtraction of two vector

```python
vec_1 = np.asarray([0,0,2,3])
vec_2 = np.asarray([0,0,5,-5])
difference = vec_1 - vec_2
print(difference)
```

```python
plt.quiver(0,0,2,3,scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2,scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-3,8,scale_units='xy', angles='xy', scale=1, color='g')

plt.xlim=(-8,8)
plt.ylim=(-8,8)
plt.show()
```

# Multiply Vector with magnitude of 2 scaler

```python
vec_1 = np.asarray([0,0,2,3])
vec_2 = 2*vec_1
print(vec_2)
```

```python
plt.quiver(0,0,2,3,scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,4,6,scale_units='xy', angles='xy', scale=1, color='r')

plt.xlim=(-8,8)
plt.ylim=(-8,8)
plt.show()
```

```python
vec_1 = np.asarray([0,0,2,3])
vec_2 = -0.5*vec_1
print(vec_2)
```

```python
plt.quiver(0,0,2,3,scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-1,-1.5,scale_units='xy', angles='xy', scale=1, color='r')

plt.xlim=(-8,8)
plt.ylim=(-8,8)
plt.show()
```
