```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data from sklearn data library
house_price_dataset = pd.read_csv('boston.csv')
house_price_dataset.head()
```

```python
# find null vlaue 
house_price_dataset.isnull().sum()
```

```python
house_price_dataset.shape
```

```python
house_price_dataset.info()
```

```python
house_price_dataset.describe()
```

```python
# find corealltion between various variables in the dataset
correl = house_price_dataset.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correl,annot=True,cmap='Blues',fmt='.1f', square=True,cbar=True)
```

```python
# Seprate the data and features
X = house_price_dataset.drop(columns='PRICE', axis=1)
Y = house_price_dataset['PRICE']
```

```python
print(X)
```

```python
print(Y)
```

## Split the data into train & test

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.1, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
print(Y.shape,Y_train.shape,Y_test.shape)
```

# Train the model

## XGBoost Regressor

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
training_data_prediction = model.predict(X_train)
training_data_prediction
```

```python
test_data_prediction = model.predict(X_test)
test_data_prediction
```

## Evalution the model 

### By checking Mean Squared ERROR & Absolute Mean Squared ERROR

```python
# R Squared ERROR
score1 = metrics.r2_score(training_data_prediction,Y_train)
# Absolute Mean ERROR
score2 = metrics.mean_absolute_error(training_data_prediction,Y_train)

print('R Squared ERROR : ', score1)
print('Absoulte ERROR : ', score2)
```

```python
# R Squared ERROR
score1 = metrics.r2_score(test_data_prediction,Y_test)
# Absolute Mean ERROR
score2 = metrics.mean_absolute_error(test_data_prediction,Y_test)

print('R Squared ERROR : ', score1)
print('Absoulte ERROR : ', score2)
```

### Make the graph to compare the actual values & predictable values

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel('Actual Prices')
plt.ylabel('Predict Prices')
plt.title('Actual Prices VS Predict Prices')
plt.show()
```

```python
plt.scatter(Y_test,test_data_prediction)
plt.xlabel('Actual Prices')
plt.ylabel('Predict Prices')
plt.title('Actual Prices VS Predict Prices')
plt.show()
```
