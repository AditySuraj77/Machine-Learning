```python
import pandas as pd
```

```python
pd.set_option('display.max_columns',None)
```

```python
credir_card = pd.read_csv('creditcard.csv')
credir_card.head()
```

```python
credir_card.Class.value_counts()
```

## 0 ===> Legal Transction
## 1 ===> Fraud Transction

```python
ligal = credir_card[credir_card.Class == 0]
fraud = credir_card[credir_card.Class == 1]
```

```python
print(ligal.shape)
print(fraud.shape)
```

#### UnderSampling  (this method is import data one class to other class)

```python
ligal_smaple = ligal.sample(n=492)
```

```python
print(ligal_smaple.shape)
```

```python
new_dataset = pd.concat([fraud,ligal_smaple], axis=0)
```
