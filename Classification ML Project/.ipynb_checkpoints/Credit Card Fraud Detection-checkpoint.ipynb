```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# 0 ---> Legit Transcation
# 1 ---> Fraud Transcation

```python
# Balance the data
legit = credit_card[credit_card['Class'] == 0]
fraud = credit_card[credit_card['Class'] == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the Class columns to another columns
credit_card.groupby('Class').mean()
```

## Under-Sampling

## Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

## Number of Fraudulent Transactions --> 492

```python
# balance the data using sample technics
legit_sample = legit.sample(n=492)
```

```python
# Concate the two dataftrames
new__data = pd.concat([legit_sample,fraud], axis=0)
```

```python
new__data.head()
```

```python
new__data.Class.value_counts()
```

```python
new__data.groupby('Class').mean()
```

## Seprate the data & label


```python
X = new__data.drop(columns='Class', axis=1)
Y = new__data['Class']
print(X)
print(Y)
```

## Split the data inti train & test form

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.1, stratify=Y,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
print(Y.shape,Y_train.shape,Y_test.shape)
```

# Train the Model

## Logistics Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

## Model Evalution

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print('Accuracy on training data = ', training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print('Accuracy on test data = ', test_data_accuracy)
```
