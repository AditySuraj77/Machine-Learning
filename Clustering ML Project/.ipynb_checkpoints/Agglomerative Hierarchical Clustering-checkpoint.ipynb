```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
customer_data = pd.read_csv('Mall_Customers.csv')
customer_data.head()
```

```python
customer_data.shape
```

```python
data = customer_data.iloc[:,3:5]
data
```

```python
print(data)
```

# Making DendoGram

```python
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(10,5))
plt.title('Customer DendoGram')
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

## Train the Model

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
clusters = AgglomerativeClustering(n_clusters=5, affinity="euclidean", linkage='ward')
```

```python
labels_ = clusters.fit_predict(data)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=clusters.labels_, cmap='rainbow')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
