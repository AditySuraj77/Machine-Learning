```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

# iris data for Encoding

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
label = LabelEncoder()
```

```python
iris['target'] = label.fit_transform(iris.variety)
```

```python
iris.target.value_counts()
```
